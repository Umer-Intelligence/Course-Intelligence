# Visualizing dataset

```python
#To find the average particpants per course
```

```python
sns.barplot(x='Course Subject',y='Participants',data=Course)
plt.title('Average Participants per Course Title',size=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#To find the percentage of males attending each course
```

```python
sns.barplot(x='Course Title',y='Per Male',data=Course.head())
plt.title('Percentage male attendence per Course subject',size=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#To find the number of courses offered by each institute
```

```python
sns.set_theme(style='dark')
sns.countplot(x='Institution',data=Course,color='red')
plt.title('Number of courses per Institute',size=15)
plt.ylabel('Number of courses')
plt.show()
```

```python
#To find average participants per subject
```

```python
Pie=Course.groupby('Course Subject')['Participants'].mean()
Pie.plot.pie(autopct="%.1f%%")
plt.title('Average participants per subject',size=15)
plt.show()
```

```python
#To find total course hours per subject
```

```python
sns.lineplot(x='Course Title',y='Total Course Hours',data=Course.head(10))
plt.xticks(rotation='vertical')
plt.title('Total course hours per subject',size=15)
plt.show()
```
